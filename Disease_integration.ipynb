```python
def inter_disease(disease_combine_all):
    disease_combine_all2 = disease_combine_all.T
    disease_combine_all2.iloc[:][:1][list(set(list(match_disease_gene['clean_name'].values)))] = 1
    disease_combine_all2.iloc[:][1:2][list(set(list(match_disease_mirna['clean_name'].values)))] = 1
    disease_combine_all2.iloc[:][2:3][list(set(list(match_disease_similarity['clean_name'].values)))] = 1
    return disease_combine_all2
asso_disease_combine_all = inter_disease(asso_disease_combine_all)
asso_disease_combine_all.iloc[:][3:4][list(set(list(asso_disease_food['clean_name'].values)))] = 1
trea_disease_combine_all = inter_disease(trea_disease_combine_all)
trea_disease_combine_all.iloc[:][3:4][list(set(list(trea_disease_food['clean_name'].values)))] = 1
```
